# Tarea Regresión Lineal Multiple

```python
# Importación de librerias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Importación de data set diabetes

from sklearn.datasets import load_diabetes

df = load_diabetes()
df
```

```python
plt.figure(figsize=(20,10))

plt.plot(df.target)
```

### Creación del modelo de regresion multiple

```python
df.feature_names
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = pd.DataFrame(df.data, columns=df.feature_names)
Y = df.target

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)

model_lin_reg_mult = LinearRegression()

model_lin_reg_mult.fit(X_train, Y_train)
```

### Evaluación del modelo

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

Y_pred = model_lin_reg_mult.predict(X_test)

mse = mean_squared_error(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)
r2 = r2_score(Y_test,Y_pred)

print(f'MSE: {mse}\nMAE: {mae}\nR2: {r2}')
print('*'*110)

plt.figure(figsize=(20,10))
plt.plot(Y_test)
plt.plot(Y_pred, 'r')
plt.legend(['Y_test','Y_perd'])
plt.show()
```
